1. Casos por doença ao longo do tempo

```python
from pyspark.sql.functions import sum

df = (
    df_fato_arboviroses_final
    .join(df_dim_tempo, on="id_tempo")
    .join(df_dim_doenca, on="id_doenca")
    .groupBy("ano", "mes", "doenca")
    .agg(sum("casos").alias("casos"))
    .orderBy("ano", "ordem_mes")
)

display(df)
```

2. Total de casos por doença

```python
df = (
    df_fato_arboviroses_final
    .join(df_dim_doenca, on="id_doenca")
    .groupBy("doenca")
    .agg(sum("casos").alias("total_casos"))
    .orderBy("total_casos", ascending=False)
)

display(df)
```

3. Casos por bairro (dengue espacial)

```python
df = (
    df_fato_dengue_espacial_final
    .join(df_dim_bairro, on="id_bairro")
    .groupBy("bairro")
    .count()
    .orderBy("count", ascending=False)
)

display(df)
```

4. Distribuição por sexo

```python
df = (
    df_fato_dengue_espacial_final
    .join(df_dim_sexo, on="id_sexo")
    .groupBy("sexo")
    .count()
)

display(df)
```
